Identifying numerical columns to check for outliers in numerical values

```python
# Function to count outliers using IQR
def count_outliers(data, columns):
    outlier_counts = {}
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Count outliers
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        outlier_counts[column] = len(outliers)
    return outlier_counts
```

count_outliers method is defined that counts the number of outliers in each numerical column using IQR method. By finding the upper and lower bound values and checking if any values in these columns are less than lower bound or greater than upper bound, then they are considered as outliers. The count of these outliers are returned in this method.

```python
# Count outliers
outliers_count = count_outliers(liquor_df, numerical_columns)

# Print number of outliers in each column
print("Number of outliers in each column:")
for column, count in outliers_count.items():
    print(f"{column}: {count}")
```

A new variable outliers_count stores the returned value of the method and as parameters, the dataset and the numerical columns are passed. Then the number of outliers in each column is printed. We have 13150 outliers in State bottle cost and state bottle retail. Most outliers are present in the Bottle volume(ml) column which has 65612 outliers. 10008 outliers in bottles sold, 15779 in sale, 9146 in volume sold(liters), 11830 in volume sold(gallons) column.

```python
# Function to handle outliers by capping
def cap_outliers(data, columns):
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Cap the outliers
        data[column] = data[column].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    return data
```

cap_outliers fucntion is defined. We will cap the outliers for numerical columns by identifying values that are outside the Interquartile Range (IQR). If any value is lower than the lower bound or higher than the upper bound, we will cap it at these bounds.

```python
# Apply the capping function
liquor_df = cap_outliers(liquor_df, numerical_columns)
```

The dataset and the numerical colums are passed to cap_outliers function to handle the outliers

```python
# Count outliers
outliers_count = count_outliers(liquor_df, numerical_columns)

# Print number of outliers in each column
print("Number of outliers in each column:")
for column, count in outliers_count.items():
    print(f"{column}: {count}")
```
